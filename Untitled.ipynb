```python
def drop_columns():
    """Here we are dropping some columns with value count equal to 1
    Here our function will return a dataframe without dropped column"""
    
    drop_columns=['listing_url','name']
    for i in df_list.columns:
        if len(df_list[i].value_counts())<=1:
            drop_columns.append(i)
    drop_columns.append('summary')
    return df_list.drop(drop_columns,axis=1)
df_list=drop_columns()
        
    
    
```

```python
def droping_columns_with_null():
    """Here we are dropping columns with null% greater than 70"""
    null_perc={}
    drop_columns=[]
    for i in df_list.columns:
        null_perc[i]=df_list[i].isnull().sum()/df_list.shape[0]
        if null_perc[i]>0.7:
            drop_columns.append(i)
            
    return df_list.drop(drop_columns,axis=1)
df_list=droping_columns_with_null()
```

```python
def converting_date_in_experience():

# using years as experience
    df_list['host_since']=pd.to_datetime(df_list['host_since'])
    df_list['host_since']=df_list['host_since'].apply(lambda x: x.year)
    df_list['host_since']=(max(df_list['host_since'])-df_list['host_since'])
    df_list['host_since']=df_list['host_since']/max(df_list['host_since'])
    return df_list['host_since']

df_list['host_since']=converting_date_in_experience()
df_list.head()
```

# observing price and removing few high prices
Exploring our dataset earlier (in Exploring Price) showed that though the extremely high-value rentals may sometimes be actual mansions or whatnot for rent, for the most part they're scalpers setting unrealistic valuations . These will throw off our model if we include them, by a lot.

Exploring Price showed that a price ceiling of around 650 per day was reasonable, so let's use that here.

```python
fitters = feature_table.query('price <= 650')
#Removing price with more that 650 and new data is named fitters
```

```python
fitters.shape
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.preprocessing import MinMaxScaler

from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.linear_model import LogisticRegression
import sklearn.metrics
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python

def model_fitting(fitters):    
    """Here we are going to fit our model in Gradient Boosting Regressor and we use grid search and it will return the model fitted to visualize data """
    X=fitters.drop('price', axis='columns')
    y=fitters['price']
        # Split the preprocessed dataset into train/test dataset
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
    print("Training set has {} samples.".format(X_train.shape[0]))
    print("Testing set has {} samples.".format(X_test.shape[0]))

    # Grid Search to find Best Parameter
    param_grid = {'n_estimators': [400], 
                      'learning_rate': [0.05],
                      'max_depth': [10],
                      'min_samples_split': [5],
                      'subsample': [0.7]
                     }

    forest_reg = GradientBoostingRegressor(random_state = 42)
    grid_search = GridSearchCV(forest_reg, param_grid, cv=5, scoring = 'neg_mean_squared_error', return_train_score=True)
    grid_search.fit(X_train, y_train)
    print(grid_search.best_params_)
    cvres = grid_search.cv_results_
    for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
        print(np.sqrt(-mean_score), params)
    y_pred = grid_search.predict(X_test)
    print("Error Score on Test Data: {}.".format(np.sqrt(sklearn.metrics.mean_squared_error(y_test, y_pred))))
    return grid_search ,X_train.columns
model,column=model_fitting(fitters)
```

```python
def feature_visual(grid_search,columns):
    '"this function plots top 40 features according to model"'
    feature_importances = grid_search.best_estimator_.feature_importances_
    features = sorted(zip(feature_importances, columns), reverse=True)
    attribute = []
    coefficient = []
    for feature in features:
        attribute.append(feature[1]) 
        coefficient.append(feature[0])
    plt.figure(figsize=(12,4))
    plt.bar(attribute[:40], height=coefficient[:40])
    plt.xticks(rotation=90)
    plt.xlabel('feature')
    plt.ylabel('feature importance')
    plt.title('feature importance for the Top 40 features')
    plt.show()

    
feature_visual(model,column)
```

```python
def least_important_visual(grid_search,columns):
     '"this function plots least iportant features according to model"'
    feature_importances = grid_search.best_estimator_.feature_importances_
    features = sorted(zip(feature_importances, columns), reverse=True)
    attribute = []
    coefficient = []
    for feature in features:
        attribute.append(feature[1]) 
        coefficient.append(feature[0])
    plt.figure(figsize=(12,4))
    plt.bar(attribute[120:], height=coefficient[120:])
    plt.xticks(rotation=90)
    plt.xlabel('feature')
    plt.ylabel('feature importance')
    plt.title('least important features')
    plt.show()
least_important_visual(model,column)  
    
```

```python

# Plot Amenity important features
def important_amenities_visual(grid_search,columns,amenities_list):
     '"this function plots top amenities according to model"'
    feature_importances = grid_search.best_estimator_.feature_importances_
    features = sorted(zip(feature_importances, columns), reverse=True)
    amenity= []
    amn_coef = []
    for feature in features:
        if feature[1] in amenities_list:
            amenity.append(feature[1])
            amn_coef.append(feature[0])
    plt.figure(figsize=(8,4))
    plt.bar(amenity[:], height=amn_coef[:])
    plt.xticks(rotation = 90)
    plt.xlabel('amenity')
    plt.ylabel('amenity importance')
    plt.title('feature importance for the Top amenities')
    plt.show()
important_amenities_visual(model,column,amenities_list)
```
